**Q6.) Yes, the data is almost certainly full of confounding variables, which are hidden factors that influence the data and can lead you to a false conclusion. The most likely confounding variable here is population density: mask mandates were most likely implemented in big, dense, urban cities where the virus spreads easily, while "no mandate" counties were likely more rural and naturally spread out. Therefore, this graph is probably just comparing "dense cities" vs. "rural towns." Another major confounding factor is reverse causality: counties issue mask mandates because their case counts are already high or spiking; the high case count caused the mandate, not the other way around. Because of this, the only truly accurate, honest statement you can make is: "During the period of July 12th to August 3rd, 2020, the group of Kansas counties with mask mandates had a higher 7 day rolling average of new COVID-19 cases per 100k than the group of counties without mask mandates."
