**ONLINE ARTICLE: Thorbecke, Catherine, Google hit with lawsuit alleging it stole data from millions of users to train its AI tools, CNN, 10/31/2025, https://www.cnn.com/2023/07/11/tech/google-ai-lawsuit**

This article focused on a 2023 lawsuit against google that claimed it was using users data to train their AI models without direct permission from the user. One of the quotes that specifically stood out to me, "Google needs to understand that ‘publicly available’ has never meant free to use for any purpose,” Tim Giordano, one of the attorneys at Clarkson bringing the suit against Google." This quote represents the main debacle when it comes to data ethics, what does it mean to be public or pivate, especially in the age of technology since information is so wide spread. This abstract question has a directly abstract answer, as it really varies from person to person on whether or not they approve their data to be utilized for someting such as AI. Google's response to this was likewise intresting to break down, since Google has been prone to data breaches in the past, yet "Google’s general counsel, called the claims in the suit 'baseless'." Me personally, I do not believe the concerns regarding how my data is being used especially for a company of Google's scale, to be groundless. This article is a nice introductury to what may be a deeper dive into data privacy, as it really presents us as the reader with a scenario, where these tools we use in our everday lives, may have some nefarious things going on in the background. As of course, nothing is ever free (even when using Google).

Honestly this article made me do a double take on all the tools I integrate into my workflow from these big tech companies, and how they may or may not violate my data privcy. For example after I read this article I checked the terms and services of OpenAI shocked to find that when you delete your chat logs, you are not actually deleting your chat logs since OpenAI's ChatGBT models operate off of memory so it will store the data from these previous chats and train the model on your data even if you delete the chat logs. With this newfound (and unsettling) information, I deleted my OpenAI account permanently. The general idea of data privacy is one that intrests me and from my exploration came a newfound distrust of companies that I previously trusted beforehand, as their ethical guidlines don't really align with my own morals.
