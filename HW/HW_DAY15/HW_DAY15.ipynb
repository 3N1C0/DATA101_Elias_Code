**Q3** Make a scatter plot of the width vs the height like the one in the lecture. You should be able to recreate the plots here without looking at the code.

The plot below uses Ordinary Least Squares fitting to find a reasonable line.

````python
DF['Height_in'] = DF['Height_in'].apply(lambda x: float(x))
DF['Width_in'] = DF['Width_in'].apply(lambda x: float(x))

fig_3 = px.scatter(DF,
                 x='Width_in',
                 y="Height_in",
                 color_discrete_sequence=['black'],
                trendline='ols',
                trendline_scope='overall',
                trendline_color_override='blue')


fig_3.update_layout(template="ggplot2",
                  title='Height vs Width of Paris Paintings <br><sup> Paris auctions, 1764-1780</sup>',
                  title_x=0.5,
                  xaxis_title="Width (inches)",
                  yaxis_title="Height (inches)")

fig_3.show()
````

So the line that "fits" this data based on the code we ran is

$$ H = 0.7808 W + 3.6214 $$

**Q4** Where do we think this prediction is most accurate? Where is there the most error? Explain why you think this?

**The prediction is most accurate in the cluster of data from the range 0 < x < 50, the most error will be at the outliers beyond this point that are farther away (in the x-direction) from the cluster. This is because the points from 0 < x < 50 are more compact, and as you move further along the x-axis the points become more polarized and tend to have more variation.**

````python
DF['landsALL'] = DF['landsALL'].apply(lambda x: str(x))

fig_4 = px.scatter(DF,
                 x='Width_in',
                 y="Height_in",
                 color='landsALL',
                 opacity=0.2,
                 trendline='ols')

fig_4.update_layout(template="ggplot2",
                  title='Height vs Width of Paris Paintings <br><sup> Paris auctions, 1764-1780</sup>',
                  title_x=0.5,
                  xaxis_title="Width (inches)",
                  yaxis_title="Height (inches)")

fig_4.show()
````

````python
# A new packages to import!
from sklearn.linear_model import LinearRegression 
from sklearn.preprocessing import OneHotEncoder
````

````python
# Select the variables - I want to predict height given width
DF_model = DF[['Height_in','Width_in']]
DF_model.dtypes
````

````python
print('Number of NaNs:')
print(DF_model.isna().sum().sum())
print('----------------------')

print('Percent NaNs:')
print(DF_model.isna().sum().sum()/len(DF))
print('----------------------')

DF_model.dropna(inplace=True)
print('Number of NaNs after drop:')
print(DF_model.isna().sum().sum())
print('----------------------')
````

````python
X = DF_model['Width_in'].values.reshape(-1, 1)
y = DF_model['Height_in'].values
````

````python
# Create linear regression object - a random straight line
LM = LinearRegression()

# Train the model using the data
LM.fit(X, y)
````

````python
LM.coef_
````

````python
LM.intercept_
````

````python
LM.score(X, y)
````

````python
fig = px.scatter(DF,
                 x='Width_in',
                 y="Height_in",
                 color_discrete_sequence=['black'],
                 opacity=0.2)

fig.update_layout(template="ggplot2",
                  title='Height vs Width of Paris Paintings <br><sup> Paris auctions, 1764-1780</sup>',
                  title_x=0.5,
                  xaxis_title="Width (inches)",
                  yaxis_title="Height (inches)")

# Add the line our model would predict
DF_model = pd.DataFrame()
DF_model['w'] = DF['Width_in']
DF_model['y'] = LM.coef_*DF_model['w']+LM.intercept_
DF_model = DF_model.sort_values('w')

fig.add_trace(
    px.line(DF_model, x='w', y='y',color_discrete_sequence=['blue']).data[0]
)

fig.show()
````

````python
DF_2 = DF[['Surface','price']]
DF_2.dtypes
````

````python
DF_2['Surface'] = pd.to_numeric(DF_2['Surface'], errors='coerce')
````

````python
print('Number of NaNs:')
print(DF_2.isna().sum().sum())
print('----------------------')

print('Percent NaNs:')
print(DF_2.isna().sum().sum()/len(DF))
print('----------------------')

DF_2.dropna(inplace=True)
print('Number of NaNs after drop:')
print(DF_2.isna().sum().sum())
print('----------------------')
````

````python
X_2 = DF_2['Surface'].values.reshape(-1, 1)
y_2 = DF_2['price'].values
````

````python
# Create linear regression object - a random straight line
LM_2 = LinearRegression()

# Train the model using the data
LM_2.fit(X_2, y_2)
````

````python
LM_2.coef_
````

````python
LM_2.intercept_
````

````python
LM_2.score(X_2,y_2)
````

````python
fig_2 = px.scatter(DF_2,  
                   x='Surface',
                   y='price',
                   color_discrete_sequence=['black'],
                   opacity=0.2)

fig_2.update_layout(template="ggplot2",
                    title='Surface vs Price of Paris Paintings <br><sup> Paris auctions, 1764-1780</sup>',
                    title_x=0.5,
                    xaxis_title="Surface (inches^2)",
                    yaxis_title="Price (U.S. Dollars)")

df_line = pd.DataFrame()

df_line['Surface'] = DF_2['Surface']

df_line['Prediction'] = LM_2.coef_[0] * df_line['Surface'] + LM_2.intercept_

df_line = df_line.sort_values('Surface')

fig_2.add_trace(
    px.line(df_line, x='Surface', y='Prediction', color_discrete_sequence=['blue']).data[0]
)

fig_2.show()
````
