# Project Proposal

### - What is the question you hope to answer?

My core research question is: "Can we build a machine learning model that accurately predicts the direction (increase or decrease) of Bitcoin's price tomorrow by combining traditional historical price data with a novel public-interest proxy (Wikipedia page edits)?"

### - Why is this question important to you? 

This question is important because it tackles the classic, difficult problem of predicting a highly volatile asset. Most simple models fail. This project tests a specific, creative hypothesis: that public interest and "buzz," as measured by Wikipedia edits, might be a valuable predictive signal that traditional price-only models miss. It's a chance to apply data science to a real-world financial puzzle and see if a unique data source can provide a genuine predictive edge.

### - What are you hoping to achieve to the project?

My primary goal is to build a functional and robust prediction system.

To achieve this, I will:

    Iterate on Models: Start by building a Random Forest model as a baseline and then improve upon it using a more powerful XGBoost model.

    Develop a Backtesting System: Create a reliable system to test the model's historical performance using a strong error metric, proving it actually performs well.

    Produce a Next-Day Predictor: The final result will be a trained model capable of ingesting new data to make a tangible, next-day prediction.
